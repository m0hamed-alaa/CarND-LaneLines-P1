## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math
from collections import deque 

# declaring a fixed_length FIFO buffer used for averaging over given number of frames

n_frames =  8            
buffer = deque( maxlen = n_frames )
buffer.clear()

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    `vertices` should be a numpy array of integer points.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

def get_x_linear(y , p) :
    """
    helper function added by me 
    It accepts the y_corrdinate and coefficients of linear model and returns the x_coordinate
    inputs :
    y - y_coordinate
    p - list containing coefficients of the linear model ex: p =[m,b]
    output :
    x - x_coordinate
    """
    x = int( ( y - p[1] ) / p[0] )
    
    return x

def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    # Detected line segments are filtered and classified into right lane line segments and left lane line segments based on slope
    
    rightLines_x1 = []
    rightLines_y1 = []
    rightLines_x2 = []
    rightLines_y2 = []

    leftLines_x1 = []
    leftLines_y1 = []
    leftLines_x2 = []
    leftLines_y2 = []

    for line in lines:
        for x1,y1,x2,y2 in line:
            slope = (y2 - y1)/(x2 - x1)
            if 0.4 < slope < 5.5  :
                rightLines_x1 += [x1]
                rightLines_y1 += [y1]
                rightLines_x2 += [x2]
                rightLines_y2 += [y2]
            elif -1.7 < slope < -0.36 :
                leftLines_x1 += [x1]
                leftLines_y1 += [y1]
                leftLines_x2 += [x2]
                leftLines_y2 += [y2]
                
    # Averaging the endpoints of the detected right and left lane line segments
    
    # Right lane lines
    
    righ_x1s_average = np.mean(rightLines_x1)
    righ_y1s_average = np.mean(rightLines_y1)
    righ_x2s_average = np.mean(rightLines_x2)
    righ_y2s_average = np.mean(rightLines_y2)
    
    # Left lane lines
    
    left_x1s_average = np.mean(leftLines_x1)
    left_y1s_average = np.mean(leftLines_y1)
    left_x2s_average = np.mean(leftLines_x2)
    left_y2s_average = np.mean(leftLines_y2)
    
    
    # Fitting a straight line model (y = m*x +b) to  the end points of the averaged right lane line segment.
    
    line_right = np.polyfit([ righ_x1s_average , righ_x2s_average ] , [ righ_y1s_average , righ_y2s_average ] , 1) # line_right = [m1,b1]

    # Fitting a straight line model (y = m*x +b) to the end points of the averaged  left lane line segment.
    
    line_left = np.polyfit([ left_x1s_average , left_x2s_average ] , [ left_y1s_average , left_y2s_average ] , 1)  # line_left  = [m2,b2]
    
    # Extrapolation 
    
    y1 = 320                  # y1 => top
    y2 = img.shape[1]         # y2 => bottom
    
    # Vertices are top right , bottom right , top left , bottom left  | vertices = [ (x1,y1) , (x2,y2) , (x1,y1) , (x2,y2) ]
    
    vertices = [ (get_x_linear(y1 , line_right ),y1) , (get_x_linear(y2 , line_right ),y2) , (get_x_linear(y1 , line_left ), y1) , (get_x_linear(y2 , line_left ), y2) ]
    
    # Drawing lane lines
    
    cv2.line(img , vertices[0] , vertices[1] , color , thickness)
    cv2.line(img , vertices[2] , vertices[3] , color , thickness)
    
    
           

def draw_lines_improved(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    
    # Detected line segments are filtered and classified into right lane line segments and left lane line segments based on slope
    
    rightLines_x = []             # to store x coordinates (x1,x2) for right lane line segments             
    rightLines_y = []             # to store y coordinates (y1,y2) for left  lane line swgments
    leftLines_x = []              # to store x coordinates (x1,x2) for right lane line segments 
    leftLines_y = []              # to store y coordinates (y1,y2) for left  lane line segments
    
    for line in lines:
        for x1,y1,x2,y2 in line:
            slope = (y2 - y1)/(x2 - x1)
            if 0.4 < slope < 5.5  :         # right lane line segments have poistive slope
                rightLines_x += [x1,x2]
                rightLines_y += [y1,y2]
            elif -1.7 < slope < -0.36 :     # left lane line segments have negative slope
                leftLines_x += [x1,x2]
                leftLines_y += [y1,y2]
                
    
    # Fitting a straight line model (y = m*x+b) to the end points of the right lane line segments
    
    line_right = np.polyfit(rightLines_x , rightLines_y , 1)     # line_right = [m1,b1]
    
    # Fitting a straight line model (y= m*x+b) to the end points of the left  lane line segments
    
    line_left = np.polyfit(leftLines_x , leftLines_y , 1)        # line_left = [m2,b2]
       
    # Keep tracking of the slopes and intercepts of the right and left lines detected and fitted in each frame 
    
    frame_info = (line_right[0],line_right[1],line_left[0],line_left[1])   # frame_info = ( m1,b1,m2,b2 )
    buffer.append(frame_info)
    
    # Taking average of m1,b1,m2,b2 which are stored in the buffer over the last 'n_frames' frames (moving average)
    
    m1_mean = 0
    b1_mean = 0
    m2_mean = 0
    b2_mean = 0
    
    for m1,b1,m2,b2 in buffer :
        m1_mean = m1_mean + m1
        b1_mean = b1_mean + b1
        m2_mean = m2_mean + m2
        b2_mean = b2_mean + b2
    
    frames_length = len(buffer)
    
    m1_mean = m1_mean / frames_length
    b1_mean = b1_mean / frames_length
    m2_mean = m2_mean / frames_length
    b2_mean = b2_mean / frames_length
    
    
    
    
    # Extrapolation 
    
    y1 = 320                  # y1 => top
    y2 = img.shape[1]         # y2 => bottom
    
    # Vertices are top right , bottom right , bottom left , top left
    
    vertices = np.array( ( (get_x_linear(y1 , [m1_mean , b1_mean] ),y1) , (get_x_linear(y2 , [m1_mean , b1_mean] ),y2) , (get_x_linear(y2 , [m2_mean , b2_mean] ), y2) , (get_x_linear(y1 , [m2_mean , b2_mean] ), y1) ) , dtype=np.int32)
    
    # Filling pixels inside the polygon defined by 'vertices' with the fill color which is green
    
    cv2.fillPoly(img , [vertices] , [0,100,0])
    
    # Drawing lane lines
    
    cv2.line(img , tuple(vertices[0]) , tuple(vertices[1]) , color , thickness)
    cv2.line(img , tuple(vertices[3]) , tuple(vertices[2]) , color , thickness)


    
    
'''    
def draw_curves(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    # Detected line segments are filtered and classified into right lane line segments and left lane line segments based on slope
    
    rightLines_x = []             # to store x coordinates (x1,x2) for right lane line segments             
    rightLines_y = []             # to store y coordinates (y1,y2) for left  lane line swgments
    leftLines_x = []              # to store x coordinates (x1,x2) for right lane line segments 
    leftLines_y = []              # to store y coordinates (y1,y2) for left  lane line segments
    
    for line in lines:
        for x1,y1,x2,y2 in line:
            slope = (y2 - y1)/(x2 - x1)
            if 0.4 < slope < 5.5  :         # right lane line segments have poistive slope
                rightLines_x += [x1,x2]
                rightLines_y += [y1,y2]
            elif -1.7 < slope < -0.36 :     # left lane line segments have negative slope
                leftLines_x += [x1,x2]
                leftLines_y += [y1,y2]
                
    # Fitting a 2nd degree polynomial model ( y = a*x^2 + b*x + c ) to the end points of the right lane line segments
    
    line_right = np.polyfit(rightLines_x , rightLines_y , 2 )     # line_right = [a1,b1,c1]
    
    # Fitting a 2nd degree polynomial model ( y = a*x^2 + b*x + c ) to the end points of the left  lane line segments
    
    line_left =  np.polyfit(leftLines_x  , leftLines_y , 2 )      # line_left = [a2,b2,c2]
    
    # Keep tracking of coefficients of the polynomials of the right and left curved lines detected and fitted in each frame 
    
    frame_info = (line_right[0],line_right[1],line_right[2],line_left[0],line_left[1],line_left[2])   
    buffer.append(frame_info)       # frame_info = ( a1,b1,c1,a2,b2,c2)
    
    # Taking average of a1,b1,c1,a2,b2,c2 which are stored in the buffer over the last 'n_frames' frames (moving average)
    
    a1_mean = 0
    b1_mean = 0
    c1_mean = 0
    a2_mean = 0
    b2_mean = 0
    c2_mean = 0
    
    for a1,b1,c1,a2,b2,c2 in buffer :
        a1_mean = a1_mean + a1
        b1_mean = b1_mean + b1
        c1_mean = c1_mean + c1
        a2_mean = a2_mean + a2
        b2_mean = b2_mean + b2
        c2_mean = c2_mean + c2
    
    frames_length = len(buffer)
    
    a1_mean = a1_mean / frames_length
    b1_mean = b1_mean / frames_length
    c1_mean = c1_mean / frames_length
    a2_mean = a2_mean / frames_length
    b2_mean = b2_mean / frames_length
    c2_mean = c2_mean / frames_length
    

    
    # Extrapolation 
    
    y1 =int(0.5*img.shape[0]+60)                # y1 => top
    y2 = img.shape[0]                           # y2 => bottom
    
    # corners are top right , bottom right , bottom left , top left
    .
    .
    .
        
    
    
    
'''

            

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines , thickness=10)
    return line_img


# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., γ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + γ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, γ)
````

````python
# TODO: Build your pipeline that will draw lane lines on the test_images
# then save them to the test_images_output directory.



def lane_finding(image , show = True , save = False) :
    
    if show == True :
        plt.figure()
        plt.title('Original RGB image')
        plt.imshow(image)

        # Print some stats about the image
        print('image dimensions are :' + str(image.shape))
        print('image size is : ' + str(image.size) )

    # Conversion to grayscale image
    gray_img = grayscale(image)
    if show == True :
        plt.figure()
        plt.title('Grayscale image')
        plt.imshow( gray_img , cmap='gray')

    # Gaussian smoothing 
    kernel_size = 5
    blurred_img = gaussian_blur( gray_img , kernel_size )
    if show == True :
        plt.figure()
        plt.title('Gaussian smoothing ')
        plt.imshow(blurred_img , cmap='gray')

    #Canny edge detection
    low_threshold = 50
    high_threshold = 150
    edges = canny( blurred_img , low_threshold , high_threshold )
    if show == True :
        plt.figure()
        plt.title('Canny edge detection')
        plt.imshow(edges , cmap='gray')

    # region of interest selection
    vertices = np.array( [[ (50,image.shape[0]) , (450,310) , (510,310) , (910,image.shape[0]) ]] , dtype = np.int32)
    masked_img = region_of_interest( edges , vertices )
    if show == True :
        plt.figure()
        plt.title('Region of interest selection')
        plt.imshow(masked_img , cmap='Greys_r')

    # Hough transform for detecting lines
    rho = 1
    theta = np.pi/180.0
    threshold = 10
    min_line_length = 10
    max_line_gap = 1
    lines_img = hough_lines( masked_img , rho ,theta , threshold , min_line_length , max_line_gap )
    if show == True :
        plt.figure()
        plt.title('Hough transform for detecting lines')
        plt.imshow(lines_img)

    # Blending the two images and show the final image
    final_img = weighted_img(lines_img , image , 0.8 , 1 , 0 )
    if show == True :
        plt.figure()
        plt.title('Final image')
        plt.imshow(final_img)
        
    return final_img
        
        
# Testing the lane finding function (pipeline)

for original_image , image_name in images :
    output_image = lane_finding( original_image , show = False , save = False )
    plt.figure()
    plt.subplot(211)
    plt.title(image_name)
    plt.imshow(original_image)
    plt.subplot(212)
    plt.title(image_name+'After processing')
    plt.imshow(output_image)
    cv2.imwrite('test_images_output/'+image_name , cv2.cvtColor(output_image , cv2.COLOR_RGB2BGR))
    
    
````

````python
n_frames =  8            
buffer = deque( maxlen = n_frames )
buffer.clear()

def process_image(image):
    # NOTE: The output you return should be a color image (3 channel) for processing video below
    # TODO: put your pipeline here,
    # you should return the final output (image where lines are drawn on lanes)
    
    
    # Conversion to grayscale image
    gray_img = grayscale(image)
    
    # Gaussian smoothing 
    kernel_size = 5
    blurred_img = gaussian_blur( gray_img , kernel_size )
    
    #Canny edge detection
    low_threshold = 50
    high_threshold = 150
    edges = canny( blurred_img , low_threshold , high_threshold )
    
    # region of interest selection
    vertices = np.array( [[ (50,image.shape[0]) , (450,310) , (510,310) , (910,image.shape[0]) ]] , dtype = np.int32)
    masked_img = region_of_interest( edges , vertices )
    
    # Hough transform for detecting lines
    rho = 1
    theta = np.pi/180.0
    threshold = 10
    min_line_length = 10
    max_line_gap = 1
    lines = cv2.HoughLinesP(masked_img, rho, theta, threshold, np.array([]), minLineLength=min_line_length, maxLineGap=max_line_gap)
    lines_img = np.zeros((image.shape[0], image.shape[1], 3), dtype=np.uint8)
    draw_lines_improved(lines_img, lines , thickness=10)
    
    
    # Blending the two images and show the final image
    result = weighted_img(lines_img , image , 0.8 , 1 , 0 )
    
    return result

def draw_lines_improved_2(img, lines, color=[255, 0, 0], thickness=2 ,roi_height=500):  #For challenge video
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    
    # Detected line segments are filtered and classified into right lane line segments and left lane line segments based on slope
    
    rightLines_x = []             # to store x coordinates (x1,x2) for right lane line segments             
    rightLines_y = []             # to store y coordinates (y1,y2) for left  lane line swgments
    leftLines_x = []              # to store x coordinates (x1,x2) for right lane line segments 
    leftLines_y = []              # to store y coordinates (y1,y2) for left  lane line segments
    
    for line in lines:
        for x1,y1,x2,y2 in line:
            slope = (y2 - y1)/(x2 - x1)
            if 0.4 < slope < 5.5  :         # right lane line segments have poistive slope
                rightLines_x += [x1,x2]
                rightLines_y += [y1,y2]
            elif -1.7 < slope < -0.36 :     # left lane line segments have negative slope
                leftLines_x += [x1,x2]
                leftLines_y += [y1,y2]
                
    
    # Fitting a straight line model (y = m*x+b) to the end points of the right lane line segments
    
    line_right = np.polyfit(rightLines_x , rightLines_y , 1)     # line_right = [m1,b1]
    
    # Fitting a straight line model (y= m*x+b) to the end points of the left  lane line segments
    
    line_left = np.polyfit(leftLines_x , leftLines_y , 1)        # line_left = [m2,b2]
       
    # Keep tracking of the slopes and intercepts of the right and left lines detected and fitted in each frame 
    
    frame_info = (line_right[0],line_right[1],line_left[0],line_left[1])   # frame_info = ( m1,b1,m2,b2 )
    buffer.append(frame_info)
    
    # Taking average of m1,b1,m2,b2 which are stored in the buffer over the last 'n_frames' frames (moving average)
    
    m1_mean = 0
    b1_mean = 0
    m2_mean = 0
    b2_mean = 0
    
    for m1,b1,m2,b2 in buffer :
        m1_mean = m1_mean + m1
        b1_mean = b1_mean + b1
        m2_mean = m2_mean + m2
        b2_mean = b2_mean + b2
    
    frames_length = len(buffer)
    
    m1_mean = m1_mean / frames_length
    b1_mean = b1_mean / frames_length
    m2_mean = m2_mean / frames_length
    b2_mean = b2_mean / frames_length
    
    
    # select the solid lane
    
    if len(rightLines_y) > len(leftLines_y) :
        solid_lane = rightLines_y 
    else :
        solid_lane = leftLines_y
    
    # Extrapolation 
    
    y1 = roi_height + 25               # y1 => top
    y2 = max(solid_lane)                # y2 => bottom
    
    # Vertices are top right , bottom right , bottom left , top left
    
    vertices = np.array( ( (get_x_linear(y1 , [m1_mean , b1_mean] ),y1) , (get_x_linear(y2 , [m1_mean , b1_mean] ),y2) , (get_x_linear(y2 , [m2_mean , b2_mean] ), y2) , (get_x_linear(y1 , [m2_mean , b2_mean] ), y1) ) , dtype=np.int32)
    
    # Filling pixels inside the polygon defined by 'vertices' with the fill color which is green
    
    #cv2.fillPoly(img , [vertices] , [0,100,0])
    
    # Drawing lane lines
    
    cv2.line(img , tuple(vertices[0]) , tuple(vertices[1]) , color , thickness)
    cv2.line(img , tuple(vertices[2]) , tuple(vertices[3]) , color , thickness)

def process_image_2(image):   # For challenge video
    
    # NOTE: The output you return should be a color image (3 channel) for processing video below
    # TODO: put your pipeline here,
    # you should return the final output (image where lines are drawn on lanes)
    
    # Conversion to HSL space

    hsl_img = cv2.cvtColor( image , cv2.COLOR_RGB2HLS)

    # Isolating white and yellow lane lines 

        # white color mask
    lower = np.uint8([  0, 200,   0])
    upper = np.uint8([255, 255, 255])
    white_mask = cv2.inRange(hsl_img , lower, upper)

        # yellow color mask
    lower = np.uint8([ 10,   0, 100])
    upper = np.uint8([ 40, 255, 255])
    yellow_mask = cv2.inRange(hsl_img , lower, upper)
   
        # combine the mask
    mask = cv2.bitwise_or(white_mask, yellow_mask)
    color_masked_img = cv2.bitwise_and(image, image, mask = mask)


    # Conversion to grayscale image

    gray_img = grayscale(color_masked_img)

    # Gaussian smoothing 

    kernel_size = 5
    blurred_img = gaussian_blur( gray_img , kernel_size )

    # Canny edge detection

    low_threshold = 20
    high_threshold = 80
    edges = canny( blurred_img , low_threshold , high_threshold )


    # region of interest selection

    top_left     = ( int(0.5*image.shape[1]-90) , int(0.5*image.shape[0]+60) )
    bottom_left  = ( int(0.15*image.shape[1]) , image.shape[0] )
    top_right    = ( int(0.5*image.shape[1]+100) , int(0.5*image.shape[0]+60) )
    bottom_right = ( int(0.85*image.shape[1]) , image.shape[0] )
    vertices = np.array( [[ bottom_left , top_left , top_right , bottom_right ]] , dtype = np.int32)
    roi_masked_img = region_of_interest( edges , vertices )


    # Hough transform for detecting lines

    rho = 1
    theta = np.pi/180.0
    threshold = 10
    min_line_length = 10
    max_line_gap = 1
    lines = cv2.HoughLinesP(roi_masked_img, rho, theta, threshold, np.array([]), minLineLength=min_line_length, maxLineGap=max_line_gap)
    lines_img = np.zeros((image.shape[0], image.shape[1], 3), dtype=np.uint8)
    draw_lines_improved_2(lines_img, lines , thickness=10 , roi_height=top_right[1])
    

    # Blending the two images and show the final image

    result = weighted_img(lines_img , image , 0.8 , 1 , 0 )
    
    return result
````
